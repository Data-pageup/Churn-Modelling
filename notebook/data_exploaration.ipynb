```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=  pd.read_csv('../data/processed/cleaned_data.csv')
```

```python
# age distribution       # UNIVARIATE 
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins = 20, kde =True)
plt.title('age distribution')
plt.show()
```

```python
# balance distribution
plt.figure(figsize=(8,5))
sns.histplot(df['balance'], bins =20, kde = True)
plt.title('balance distribution')
plt.show()   # almost 3500 people are near 0 balance !
```

```python
plt.figure(figsize=(8,5))
ax = sns.countplot(x = 'gender',data = df)
ax.bar_label(ax.containers[0])
plt.title('gender distribution')
plt.show()
```

```python
plt.figure(figsize= (8,5))
ax = sns.countplot(x = 'exited',data = df)
ax.bar_label(ax.containers[0])
plt.title('churn vs stay')   # the dataset is quite imbalance 
```

```python
plt.figure(figsize = (8,5))
ax = sns.countplot(x = 'geography', data = df)
ax.bar_label(ax.containers[0])
plt.title('geography')
plt.show()
```

```python
# now bivariate analysis

plt.figure(figsize = (8,5))
sns.scatterplot(x = 'age', y = 'creditscore' ,hue = 'exited', data = df)
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x = 'gender', hue = 'exited',data = df)
```

```python
plt.figure(figsize = (8,5))
ax = sns.countplot(x = 'geography', hue = 'exited', data =df )
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[['creditscore','age','tenure',
       'balance', 'numofproducts', 'hascrcard', 'isactivemember',
       'estimatedsalary', 'exited']].corr(),annot = True,cmap = 'coolwarm')
plt.title('correlation heatmap')
plt.show()     # only int and float dtype 
```

```python
Categorical vs Target (exited)

sns.countplot(x='Gender', hue='Exited', data=df) → churn by gender

sns.countplot(x='Geography', hue='Exited', data=df) → churn by geography

sns.countplot(x='NumOfProducts', hue='Exited', data=df) → churn by number of products

Numerical vs Target (exited)

sns.boxplot(x='Exited', y='CreditScore', data=df) → credit score distribution for churn vs non‑churn

sns.boxplot(x='Exited', y='Age', data=df) → age distribution by churn

sns.boxplot(x='Exited', y='Balance', data=df) → balance differences

Numerical vs Numerical

sns.scatterplot(x='Age', y='CreditScore', hue='Exited', data=df) → age vs credit score, colored by churn

sns.scatterplot(x='Balance', y='EstimatedSalary', hue='Exited', data=df) → balance vs salary, churn highlighted

sns.heatmap(df[['CreditScore','Age','Balance','EstimatedSalary']].corr(), annot=True) → correlation between numeric features
```
